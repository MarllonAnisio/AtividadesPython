Relacionais (Comparações): == (igualdade), != (diferença), >, <, >=, <=.

Lógicos: Aqui usamos palavras em vez de símbolos confusos. Usamos and, or e not


operadores basicos

```python
A = 5
B = 7
C = A + B # Soma, podems usar a soma atraves do +

D = (C + A) - B # subtração, usamos o -

X = A * B # multiplicação, podemos usar o operador *

Y = X / D # Divisão que sempre retorna floats

H = Y // C # Isso é divisão inteira

E = A ** B # isso é exponenciação

# temos o reto tmb que é %
# comparação  ==
# !=  diferença
# > maior que
# < menor que
# >= maior ou igual
# <= menor ou igual
#
# and == &&
#
# or == ||
#
# not é a negação podemos usar o !

print(f'a soma foi {C}, a subtração foi : {D} a multiplicação foi {X} a divisão foi {Y} a divisão inteira {H} e a exponenciação {E}');

```

**Listas (As famosas "Arrays" com superpoderes)**

```python
# ======================================================================
# ESTRUTURAS DE DADOS: LISTAS (Mutáveis)
# ======================================================================
# Listas servem para guardar vários itens no mesmo lugar. Usamos colchetes [].
# No Python, uma lista pode misturar números, textos e até outras listas.

salarios_equipe = [1500.50, 3200.00, 8500.75, 12000.00]

# REGRA DE OURO: A contagem sempre começa no ZERO.
# Se quiser o primeiro item, você pede o índice 0. Se pedir o 1, vem o segundo.
primeiro_salario = salarios_equipe[0]
segundo_salario = salarios_equipe[1]

# Listas são MUTÁVEIS. O mercado mudou? Você atualiza o valor direto lá dentro:
salarios_equipe[1] = 3500.00 # Aumentamos o segundo salário

# O comando mais usado do universo: append(). Ele joga um item novo no final da lista.
salarios_equipe.append(25000.00) # Chegou o diretor

# A função len() (de length/tamanho) conta quantos itens tem ali dentro.
tamanho_da_lista = len(salarios_equipe)

print(f'O primeiro salário é {primeiro_salario}.')
print(f'Agora temos {tamanho_da_lista} salários cadastrados na equipe: {salarios_equipe}')
```

Dicionários (Trabalhando com Chave e Valor)

```python
# ======================================================================
# ESTRUTURAS DE DADOS: DICIONÁRIOS
# ======================================================================
# Esqueça os índices numéricos (0, 1, 2). Aqui nós damos NOME aos bois.
# Usamos chaves {}. É o formato mais próximo de um JSON que você vai ver.
# Excelente para mapear características de um objeto ou entidade.

desenvolvedor = {
    "nome": "João Sênior",
    "linguagem_principal": "Python",
    "anos_experiencia": 5,
    "ativo": True
}

# Em vez de buscar por um número, buscamos pela chave exata (o nome do atributo):
nome_dev = desenvolvedor["nome"]
linguagem = desenvolvedor["linguagem_principal"]

# Podemos adicionar novas chaves dinamicamente a qualquer momento:
desenvolvedor["salario_alvo"] = 15000.00

print(f'O programador {nome_dev} foca em {linguagem}.')
print(f'A pretensão salarial atualizada dele é de R$ {desenvolvedor["salario_alvo"]}')
```
